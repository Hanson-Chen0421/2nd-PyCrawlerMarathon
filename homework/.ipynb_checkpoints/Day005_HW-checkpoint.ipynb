```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

response
```

```python
#é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import json
data = json.loads(response)

b = []
for a in data :
    b.append(a['title'])
    
print(len(b))
```

```python
#å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
for a in data :
    print(a['title'])
    print(a['createdAt'])
    print(a['commentCount'])
    print(a['likeCount'])
```

```python
#è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
comment_cnt = 0
like_cnt = 0
for a in data :
    comment_cnt += a['commentCount']
    like_cnt += a['likeCount']

print("ç†±é–€æ–‡ç«  å¹³å‡ç•™è¨€äººæ•¸: ", comment_cnt / len(b)) 
print("ç†±é–€æ–‡ç«  å¹³å‡æŒ‰è®šäººæ•¸", like_cnt / len(b))

```
